```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

**범례의 위치**를 설정하려며 **loc 인자**를 입력해야 합니다. loc 인자에 입력할 수 있는 값은 다음과 같습니다.

- "left"
- "center"
- "right"
- "upper left"
- "upper center"
- "upper right"
- "lower left"
- "lower center"
- "lower right"

```python
plt.scatter([1,1,1], [1,2,3], label="x=1")
plt.scatter([-1,-1,-1], [1,2,3], label="x=-1")
plt.legend(loc = "upper center")
```

```python
#씨본(seaborn) 스트일 설정
import seaborn as sns
sns.set(style = "whitegrid") # 기본값(style="darkgrid")
plt.scatter([1,2,3], [4,5,6])
```

```python
sns.set(style = "darkgrid")
plt.xlim(left = -1, right = 10)
plt.ylim(top = 11)
plt.scatter([0, 9, 2, 5], [1, 9, 9 , 8])
plt.savefig('sample_graph.png', dpi = 100)
```
