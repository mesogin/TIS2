#데이터 병합
##concat 함수 기초
concat 함수는 둘 이상의 시리즈 혹은 데이터프레임을 행 혹은 열 방향으로 이어붙이는 함수이다.

[그림: concat 함수](https://docs.google.com/presentation/d/1xS84-v_VnsHBtdBWNztR46BOskrzOVpI7re7iSlDFO0/edit?usp=sharing)


```python
#행방향 데이터 병합
import pandas as pd
df1 = pd.DataFrame({"A":[1,2,3], "B":[1,2,3]})
df2 = pd.DataFrame({"A":[4,5], "B":[4,5]})
# display(df1)
# display(df2)

#axis인자를 0으로 설정하고 ignor_index를 True와 False로 설정해서 병합하기
display(pd.concat([df1, df2], axis = 0, ignore_index = True)) #기존의 인덱스를 활용 안해도 된다면 True
display(pd.concat([df1, df2], axis = 0, ignore_index = False))  #기존의 인덱스를 그대로 활용해야 한다면 False

df3 = pd.concat([df1, df2], axis = 0, ignore_index = False)
df3.loc[0]
```

```python
#연습문제
#두개의 데이터프레임을 위아래로 연결하시오.
import numpy as np
df1 = pd.DataFrame(np.arange(12).reshape(4, 3), index=["A", "B", "C", "D"])
df2 = pd.DataFrame(np.arange(12).reshape(4, 3), index=["a", "b", "c", "d"])

#연결하기
df3 = pd.concat([df1, df2], ignore_index=False, axis =0)

display(df3)
```

```python
#열방향 데이터 병합

df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6]})
df2 = pd.DataFrame({"C":[7,8,9]})
df3 = pd.concat([df1, df2], axis = 1, ignore_index = True) #인덱스값을 디폴트값으로 가져옴
display(df3)

#df3에 인덱스값 넣기
df3.index = ["A", "B", "C"]
display(df3)
display(pd.concat([df1, df2], axis = 1, ignore_index = False)) #인덱스값을 그대로 가져옴
```

##조인 방법 결정 : how
how 인자는 조인 방법을 설정하며, 입력할 수 있는 값은 다음과 같습니다.
- **inner**(기본값): 조인 키에 대한 교집합을 기준으로 병합(왼쪽 조인키와 오른쪽 조인 키에 공통으로 포함된 요소)
- **left**: 왼쪽 데이터프레임의 조인 키를 기준으로 병합
- **right**: 오른쪽 데이터프레임의 조인 키를 기준으로 병합
-**outer**: 왼쪽, 오른쪽 데이터프레임의 조인 키에 대한 합집합 기준으로 병합

```python
#각각의 데이터프레임에서 서로 다른 이름을 가지는 행 하나씩을 제거(drop 메서드 활용)
sports = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TIS/연습파일/sports.csv")
subject = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TIS/연습파일/subject.csv")
sports.drop(index = 9, axis=0, inplace=True) #스포츠에서 채관수님 데이터 빼기
subject.drop(8, axis=0, inplace=True) #과목에서 김중교 데이터 빼기

#성명 -> 이름
subject.rename({"성명":"이름"}, axis = 1, inplace = True)
#그런 다음, 두 데이터프레임을 병합하는데, how의 4가지 방법 모두 써보기
#inner
display(pd.merge(sports, subject, on = ["번호", "이름"], how = "inner"))
display(pd.merge(sports, subject, on = ["번호", "이름"], how = "left"))
display(pd.merge(sports, subject, on = ["번호", "이름"], how = "right"))
display(pd.merge(sports, subject, on = ["번호", "이름"], how = "outer"))
```
