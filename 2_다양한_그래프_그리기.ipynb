```python
x1 = np.random.normal(10, 5, 1000)
x2 = np.random.normal(10, 10, 100)
plt.boxplot([x1, x2])
plt.show()
```

```python
#ÌûàÏä§ÌÜ†Í∑∏Îû®
#ÎèÑÏàòÎ∂ÑÌè¨ÌëúÎ•º ÏãúÍ∞ÅÌôîÌïú Í≤ÉÏúºÎ°ú Î≥ÄÏàòÏùò Î∂ÑÌè¨Î•º ÌôïÏù∏ÌïòÎäîÎç∞ Ï†ÅÏ†àÌïú ÏãúÍ∞ÅÌôî Î∞©Î≤ïÏù¥Îã§. hist()Ìï®Ïàò ÏÇ¨Ïö©
x = np.random.normal(10, 2, 100) #ÌèâÍ∑†10 Î∂ÑÌè¨2 100Í∞úÏÉùÏÑ±
plt.hist(x)
plt.show()
```

```python
#Íµ¨Í∞Ñ ÎÇòÎàÑÍ∏∞
plt.hist(x, bins = 20)
plt.show()
```

```python
#ÌûàÌä∏Îßµ
# ÏòàÏ†úÏóê ÏÇ¨Ïö©Ìï† Îç∞Ïù¥ÌÑ∞

import pandas as pd

data  = pd.DataFrame({"ÏàòÌïô":[100, 80, 90], "ÏòÅÏñ¥":[70,90,80]},
                     index=["Ï≤†Ïàò", "ÏòÅÌù¨", "ÎØºÏàò"])
display(data)

# ÌûàÌä∏Îßµ Í∑∏Î¶¨Í∏∞
import seaborn as sns
sns.heatmap(data)
```

```python
#ÎààÍ∏àÍ∞í ÏÑ§Ï†ï
#ÌñâÍ≥º Ïó¥ Ïù∏Îç±Ïä§Í∞Ä Îî∞Î°ú ÏóÜÎäî Ï§ëÏ≤©Î∞∞Ïó¥Ïù¥ ÏûÖÎ†•ÎêòÎ©¥ xÏ∂ï,yÏ∂ïÏùò ÎààÍ∏àÍ∞íÏùÑ ÏßÅÏ†ë ÏÑ§Ï†ïÌï¥ÏïºÌï®
sns.heatmap(data)
plt.xticks([0.5, 1.5], ["ÏàòÌïô", "ÏòÅÏñ¥"])
plt.yticks([0.5, 1.5, 2.5], ["Ï≤†Ïàò", "ÏòÅÌù¨", "ÎØºÏàò"])
```

```python
#Îç∞Ïù¥ÌÑ∞ ÌëúÏãú
#annot Ïù∏ÏûêÎ•º TrueÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ Î∞ïÏä§ÏôÄ ÎåÄÏùëÎêòÎäî Îç∞Ïù¥ÌÑ∞Í∞Ä ÌëúÏãúÎê©ÎãàÎã§. (fmt Ïù∏ÏûêÎ•º Îç∞Ïù¥ÌÑ∞Î•º Ïñ¥ÎñªÍ≤å ÌëúÌòÑÌï†ÏßÄ ÏÑ§Ï†ï)
sns.heatmap(data, annot = True, fmt= ".2f")  #Îç∞Ïù¥ÌÑ∞ÌëúÏãú, ÏÜåÏàòÏ†ê 2ÏûêÎ¶¨ÏàòÍπåÏßÄ
```

```python
#ÏÑ† Ïä§ÌÉÄÏùº ÏÑ§Ï†ï
#ÎëêÍªòÏôÄ ÏÉâÏÉÅÏùÄ linewidthsÏôÄ linecolor Ïù∏ÏûêÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú ÏÑ§Ï†ïÌï®
sns.heatmap(data, annot = True, fmt= ".2f", linewidth=1, linecolor="black")
```

```python
#Ïª¨Îü¨Îßµ ÏÑ§Ï†ï
#ÏÉâÏÉÅ Î≥ÄÍ≤Ω cmap Ïù∏Ïûê ÏÇ¨Ïö©
sns.heatmap(data, annot = True, fmt= ".2f", 
            cmap="Reds", linewidth=1, linecolor="white")
```
